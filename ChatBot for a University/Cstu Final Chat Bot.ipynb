```python
# Function to generate embeddings using OpenAI
def get_embeddings(texts):
    try:
        response = openai.Embedding.create(
            model="text-embedding-ada-002",
            input=texts
        )
        return [embedding["embedding"] for embedding in response['data']]
    except Exception as e:
        print(f"Error generating embeddings: {e}")
        return []
# Function to handle the chatbot conversation
def chat_complete_messages(messages, temperature=0.7):
    try:
        completion = openai.ChatCompletion.create(
        model="gpt-4o",  # Or use "gpt-4" if available for your use case
        messages=messages,
        temperature=temperature
    )
        return completion.choices[0].message['content']
    except Exception as e:
        print(f"Error completing chat message: {e}")
        return "I'm sorry, I couldn't process your request at the moment."


# Function to query Pinecone for relevant info
def get_relevant_info(query, metadata_type, top_k=3):
    query_embedding = get_embeddings([query])[0]
    query_result = index.query(
        vector=query_embedding,
        top_k=top_k,  # Number of top results to return
        include_metadata=True
    )

    # Filter results to ensure they match the metadata_type
    filtered_results = [
        match for match in query_result['matches']
        if match['metadata']['type'] == metadata_type
    ]

    return filtered_results


# Main chatbot loop (for interactive use)
print("Welcome! I'm here to assist you with information about courses, International Students FAQs, locations, and jobs.")
# Function to handle Gradio inputs and outputs
def chatbot(user_input, chatContext):
    print(f"User Input: {user_input}")
    chatContext.append({"role": "user", "content": user_input})

    # Determine the type of information the user is requesting
    metadata_type = None
    if any(keyword in user_input.lower() for keyword in ["course", "class", "module", "study"]):
        metadata_type = "Course_Details"
    elif any(keyword in user_input.lower() for keyword in ["faq", "fees", "payment", "international student"]):
        metadata_type = "International_Students_faqs"
    elif any(keyword in user_input.lower() for keyword in ["location", "place", "map", "address"]):
        metadata_type = "locations"
    elif any(keyword in user_input.lower() for keyword in ["job", "position", "work", "assistant"]):
        metadata_type = "On_Campus_Jobs"

    if metadata_type:
        # Query Pinecone for nearest neighbors based on user input
        query_results = get_relevant_info(user_input, metadata_type)

        if query_results:
            response_message_content = f"I found some relevant information about {metadata_type}:\n"
            for match in query_results:
                response_message_content += f"- {match['metadata']['text']} (from {match['metadata']['type']})\n"
        else:
            response_message_content = f"I'm sorry, I couldn't find any relevant information about {metadata_type}."
    else:
        # Use OpenAI to generate responses if the type is not recognized
        response_message_content = chat_complete_messages(chatContext)

    chatContext.append({"role": "assistant", "content": response_message_content})
    print("ChatBot: ", response_message_content)

    # Ensure the updated chatContext is returned
    return response_message_content, chatContext

# Gradio interface
with gr.Blocks() as demo:
    gr.Markdown("# CSTU Virtual Assistant")
    with gr.Row():
        with gr.Column():
            chatbot_input = gr.Textbox(lines=1, label="Your question:", placeholder="Ask anything about CSTU...")
            chatbot_output = gr.Textbox(lines=10, label="Chatbot Response:")
            chatbot_history = gr.State(initial_chat_context)  # Initialize chat history with the system message

    # Define interaction
    def respond(user_input, chatContext):
        response, updated_history = chatbot(user_input, chatContext)
        # Update the UI state with the latest chat history
        chatbot_history.value = updated_history
        return response, updated_history

    chatbot_input.submit(respond, inputs=[chatbot_input, chatbot_history], outputs=[chatbot_output, chatbot_history])

# Run the Gradio interface
demo.launch(debug=True)
```
