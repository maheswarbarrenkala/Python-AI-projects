```python
# JSON data directly defined within the code
json_data = {
    "courses": [
    {"id": "course_1", "text": "Introduction to Cloud Computing: Learn about cloud infrastructure, deployment models, and service models."},
    {"id": "course_2", "text": "Advanced Machine Learning: Deep dive into neural networks, reinforcement learning, and generative models."},
    {"id": "course_3", "text": "Introduction to Network and Security: Learn about network protocols, security detailing, etc."},

    {"id": "prof_1", "text": "Dr. John Doe, expert in Cloud Computing and Distributed Systems, teaches introductory and advanced cloud courses."},
    {"id": "prof_2", "text": "Dr. Michel, specializes in Machine Learning and Artificial Intelligence, focuses on cutting-edge ML techniques."},
    {"id": "prof_3", "text": "Dr. Glen, specializes in Machine Learning and Artificial Intelligence, focuses on cutting-edge ML techniques."}
]
,
    "international_faqs": [
  {
    "id": "faq_1",
    "question": "What are the different types of fees for international students?",
    "answer": "There are several fees associated with studying at CSTU University for international students. These include:\n* Technology Fee: $50 (one-time fee)\n* Students' Association Fee: $40 (per semester)\n* Registration Fee: $250 (one-time fee)\n* Graduation Fee: $250 (one-time fee)\n* Per Credit Fee: $700\n**Note:** Each Course has 1.5 Credits and lasts one Term (2 months).",
    "text": "What are the different types of fees for international students? There are several fees associated with studying at CSTU University for international students. These include: Technology Fee: $50 (one-time fee), Students' Association Fee: $40 (per semester), Registration Fee: $250 (one-time fee), Graduation Fee: $250 (one-time fee), Per Credit Fee: $700. Note: Each Course has 1.5 Credits and lasts one Term (2 months)."
  },
  {
    "id": "faq_2",
    "question": "How can I find out the total cost of my program?",
    "answer": "To calculate the total cost of your program, you'll need to consider:\n* The number of credits required for your program (available on the program information page).\n* The Per Credit Fee ($700).\n* Additional fees like the Registration Fee and Graduation Fee (if applicable).\nFor example, a 12-credit program would cost approximately $8,400 in tuition fees (12 credits * $700 per credit).",
    "text": "How can I find out the total cost of my program? To calculate the total cost of your program, you'll need to consider: The number of credits required for your program (available on the program information page), The Per Credit Fee ($700), Additional fees like the Registration Fee and Graduation Fee (if applicable). For example, a 12-credit program would cost approximately $8,400 in tuition fees (12 credits * $700 per credit)."
  },
  {
    "id": "faq_3",
    "question": "What payment methods are available for international students?",
    "answer": "International students can pay their fees through the following methods:\n* **Credit Card:** Accepted credit cards include Visa, Mastercard, and American Express. A secure online payment portal is available within the university system.\n* **Bank Transfer:** International students can initiate a bank transfer to the university's designated account. Details and instructions will be provided upon request.",
    "text": "What payment methods are available for international students? International students can pay their fees through the following methods: Credit Card: Accepted credit cards include Visa, Mastercard, and American Express. A secure online payment portal is available within the university system. Bank Transfer: International students can initiate a bank transfer to the university's designated account. Details and instructions will be provided upon request."
  },
  {
    "id": "faq_4",
    "question": "When are the deadlines for fee payments?",
    "answer": "Fee payment deadlines are typically set before the start of each semester. Specific deadlines will be communicated to students via email and are also available on the university website. Timely payments are crucial to avoid late fees and potential registration holds.",
    "text": "When are the deadlines for fee payments? Fee payment deadlines are typically set before the start of each semester. Specific deadlines will be communicated to students via email and are also available on the university website. Timely payments are crucial to avoid late fees and potential registration holds."
  },
  {
    "id": "faq_5",
    "question": "What happens if I have questions about my student account or need assistance with payments?",
    "answer": "The university's International Student Office is here to help! You can contact them via email at [email protected] or by phone at [phone number]. They can assist with any questions or concerns regarding your student account and fee payments.",
    "text": "What happens if I have questions about my student account or need assistance with payments? The university's International Student Office is here to help! You can contact them via email at [email protected] or by phone at [phone number]. They can assist with any questions or concerns regarding your student account and fee payments."
  },
  {
    "id": "faq_6",
    "question": "What is Curricular Practical Training (CPT)?",
    "answer": "Curricular Practical Training (CPT) is an authorization for F-1 international students to gain practical experience directly related to their major field of study. It must be an integral part of your academic program and requires approval from your Designated School Official (DSO). There are limitations on the duration of CPT, so be sure to consult the International Student Office for details and eligibility requirements.",
    "text": "What is Curricular Practical Training (CPT)? Curricular Practical Training (CPT) is an authorization for F-1 international students to gain practical experience directly related to their major field of study. It must be an integral part of your academic program and requires approval from your Designated School Official (DSO). There are limitations on the duration of CPT, so be sure to consult the International Student Office for details and eligibility requirements."
  },
  {
    "id": "faq_7",
    "question": "What is Optional Practical Training (OPT)?",
    "answer": "Optional Practical Training (OPT) allows F-1 students temporary employment authorization to gain practical experience after completing their academic studies. There are two types of OPT: pre-completion (before graduation) and post-completion (after graduation). Students must apply for OPT through USCIS with approval from their DSO. STEM (Science, Technology, Engineering, and Mathematics) program graduates may be eligible for an extended OPT period.",
    "text": "What is Optional Practical Training (OPT)? Optional Practical Training (OPT) allows F-1 students temporary employment authorization to gain practical experience after completing their academic studies. There are two types of OPT: pre-completion (before graduation) and post-completion (after graduation). Students must apply for OPT through USCIS with approval from their DSO. STEM (Science, Technology, Engineering, and Mathematics) program graduates may be eligible for an extended OPT period."
  },
  {
    "id": "faq_8",
    "question": "What is SEVIS (Student and Exchange Visitor Information System)?",
    "answer": "The Student and Exchange Visitor Information System (SEVIS) is a U.S. Department of Homeland Security (DHS) database that tracks F and M nonimmigrant students enrolled in U.S. schools. All educational institutions approved to enroll international students must report student information to SEVIS. Maintaining a valid SEVIS record is crucial for your immigration status as an international student.",
    "text": "What is SEVIS (Student and Exchange Visitor Information System)? The Student and Exchange Visitor Information System (SEVIS) is a U.S. Department of Homeland Security (DHS) database that tracks F and M nonimmigrant students enrolled in U.S. schools. All educational institutions approved to enroll international students must report student information to SEVIS. Maintaining a valid SEVIS record is crucial for your immigration status as an international student."
  },
  {
    "id": "faq_9",
    "question": "Where can I find more information about CPT, OPT, and SEVIS?",
    "answer": "We recommend consulting the following resources for more detailed information on CPT, OPT, and SEVIS:\n* CSTU University International Student Office website\n* U.S. Department of Homeland Security Study in the States website (https://studyinthestates.dhs.gov/assets/SEVP_InternationalStudentLifecycle_2022-BW.pdf)",
    "text": "Where can I find more information about CPT, OPT, and SEVIS? We recommend consulting the following resources for more detailed information on CPT, OPT, and SEVIS: CSTU University International Student Office website, U.S. Department of Homeland Security Study in the States website (https://studyinthestates.dhs.gov/assets/SEVP_InternationalStudentLifecycle_2022-BW.pdf)"
  }
]
,
    "jobs": [
  {
    "id": "job_1",
    "type": "on-campus",
    "title": "Work-Study Program",
    "description": "The Federal Work-Study program provides part-time employment opportunities for eligible students with financial need. On-campus jobs can offer flexible hours and valuable work experience related to your field of study. Examples include working in libraries, administrative offices, or student support services.",
    "eligibility": "To be eligible for the Work-Study program, you'll need to complete the Free Application for Federal Student Aid (FAFSA) and demonstrate financial need. International students with F-1 visas may be eligible under specific circumstances. Consult the International Student Office for details.",
    "application": "Applications for the Work-Study program are typically submitted through the university's financial aid office website. Once approved, you can browse job postings on the university's career portal or inquire with departments directly for potential openings.",
    "text": "Work-Study Program (On-Campus). The Federal Work-Study program provides part-time employment opportunities for eligible students with financial need. On-campus jobs can offer flexible hours and valuable work experience related to your field of study. Examples include working in libraries, administrative offices, or student support services. To be eligible for the Work-Study program, you'll need to complete the Free Application for Federal Student Aid (FAFSA) and demonstrate financial need. International students with F-1 visas may be eligible under specific circumstances. Consult the International Student Office for details. Applications for the Work-Study program are typically submitted through the university's financial aid office website. Once approved, you can browse job postings on the university's career portal or inquire with departments directly for potential openings."
  },
  {
    "id": "job_2",
    "type": "on-campus",
    "title": "Resident Assistant (RA)",
    "description": "Resident Assistants (RAs) play a vital role in fostering a positive and supportive living environment within on-campus residence halls. Responsibilities include supervising student residents, enforcing community policies, and assisting with resident needs. RAs typically receive free or subsidized housing and a meal plan as compensation.",
    "eligibility": "To be eligible for an RA position, you must be a current CSTU student with a strong academic record and excellent communication skills. A genuine interest in working with students and creating a positive community atmosphere is essential.",
    "application": "Applications for RA positions are usually announced during specific periods each semester. Look out for postings on the university housing website or inquire with the housing department for details and application deadlines.",
    "text": "Resident Assistant (RA) (On-Campus). Resident Assistants (RAs) play a vital role in fostering a positive and supportive living environment within on-campus residence halls. Responsibilities include supervising student residents, enforcing community policies, and assisting with resident needs. RAs typically receive free or subsidized housing and a meal plan as compensation. To be eligible for an RA position, you must be a current CSTU student with a strong academic record and excellent communication skills. A genuine interest in working with students and creating a positive community atmosphere is essential. Applications for RA positions are usually announced during specific periods each semester. Look out for postings on the university housing website or inquire with the housing department for details and application deadlines."
  },
  {
    "id": "job_3",
    "type": "off-campus",
    "title": "General Off-Campus Jobs",
    "description": "Numerous off-campus job opportunities exist in the surrounding area, depending on your skills and experience. Popular options include retail stores, restaurants, hospitality services, or tutoring positions. Utilize online job boards, local classifieds, or network with local businesses to explore available opportunities.",
    "eligibility": "Eligibility for off-campus jobs depends on your specific visa status and work authorization. International students with F-1 visas may be eligible for off-campus employment under CPT (Curricular Practical Training) or OPT (Optional Practical Training) programs. Consult the International Student Office for guidance on work authorization and restrictions.",
    "application": "Research and apply for jobs directly through company websites, online job boards, or by visiting businesses in person. Highlight your skills and experience relevant to the position in your resume and cover letter.",
    "text": "General Off-Campus Jobs. Numerous off-campus job opportunities exist in the surrounding area, depending on your skills and experience. Popular options include retail stores, restaurants, hospitality services, or tutoring positions. Utilize online job boards, local classifieds, or network with local businesses to explore available opportunities. Eligibility for off-campus jobs depends on your specific visa status and work authorization. International students with F-1 visas may be eligible for off-campus employment under CPT (Curricular Practical Training) or OPT (Optional Practical Training) programs. Consult the International Student Office for guidance on work authorization and restrictions. Research and apply for jobs directly through company websites, online job boards, or by visiting businesses in person. Highlight your skills and experience relevant to the position in your resume and cover letter."
  },
  {
    "id": "job_4",
    "type": "off-campus",
    "title": "Internships",
    "description": "Internships provide valuable real-world experience in your field of study and can enhance your resume. Explore internship opportunities with local companies, government agencies, or non-profit organizations related to your academic interests. The university's career center can offer guidance on internship search and application procedures.",
    "eligibility": "Eligibility for internships may vary depending on the program and company requirements. International students with F-1 visas may need specific work authorization under CPT or OPT to participate in internships. Consult the International Student Office for details.",
    "application": "Search for internship opportunities through the university's career center website, online job boards, or by directly contacting companies of interest. Prepare a compelling resume and cover letter highlighting your academic background and skills relevant to the internship position.",
    "text": "Internships (Off-Campus). Internships provide valuable real-world experience in your field of study and can enhance your resume. Explore internship opportunities with local companies, government agencies, or non-profit organizations related to your academic interests. The university's career center can offer guidance on internship search and application procedures. Eligibility for internships may vary depending on the program and company requirements. International students with F-1 visas may need specific work authorization under CPT or OPT to participate in internships. Consult the International Student Office for details. Search for internship opportunities through the university's career center website, online job boards, or by directly contacting companies of interest. Prepare a compelling resume and cover letter highlighting your academic background and skills relevant to the internship position."
  }
]
,
    "location": [
  {
    "id": "location_1",
    "text": "California Science and Technology University (CSTU), University located at 100 Innovation Way, Santa Clara, CA 95054. [Google Maps Link](https://www.google.com/maps/place/California+Science+and+Technology+University+(CSTU)++100+Innovation+Way,+Santa+Clara,+CA+95054)"
  },
  {
    "id": "location_2",
    "text": "Santa Clara Valley Medical Center, Hospital located at 751 S. Wells Fargo Dr, Santa Clara, CA 95054. [Google Maps Link](https://www.google.com/maps/place/Santa+Clara+Valley+Medical+Center,+751+S+Wells+Fargo+Dr,+Santa+Clara,+CA+95054)"
  },
  {
    "id": "location_3",
    "text": "Santa Clara Police Department - Main Station, Police Services located at 150 Santa Clara Ave, Santa Clara, CA 95050. [Google Maps Link](https://www.google.com/maps/place/Santa+Clara+Police+Department+-+Main+Station,+150+Santa+Clara+Ave,+Santa+Clara,+CA+95050)"
  },
  {
    "id": "location_4",
    "text": "Food options near CSTU (replace with your preferred categories). This location entry provides a placeholder for including various food options around CSTU."
  },
  {
    "id": "sub_location_1",
    "text": "Coffee Shop Example located at 200 Innovation Way, Santa Clara, CA 95054. [Google Maps Link](https://www.google.com/maps/place/Coffee+Shop+Example+(placeholder)+200+Innovation+Way,+Santa+Clara,+CA+95054)"
  },
  {
    "id": "sub_location_2",
    "text": "Restaurant Example located at 35 Main St, Santa Clara, CA 95050. [Google Maps Link](https://www.google.com/maps/place/Restaurant+Example+(placeholder)+35+Main+St,+Santa+Clara,+CA+95050)"
  }
]

}


# Function to generate embeddings using OpenAI
def get_embeddings(texts):
    response = openai.Embedding.create(
        model="text-embedding-ada-002",  # Use the appropriate embedding model
        input=texts
    )
    return [embedding["embedding"] for embedding in response['data']]

# Function to upsert data to Pinecone
def upsert_data(index):
    for metadata_type, data in json_data.items():
        texts = [doc['text'] for doc in data]
        ids = [doc['id'] for doc in data]

        embeddings = get_embeddings(texts)

        # Prepare vectors with metadata to identify the source
        vectors = [(doc_id, embedding, {"type": metadata_type, "text": text})
                   for doc_id, embedding, text in zip(ids, embeddings, texts)]

        # Upsert vectors to Pinecone
        index.upsert(vectors)

        print(f"Documents from {metadata_type} have been upserted into Pinecone successfully.")

# Call the function to upsert data
upsert_data(index)

```

```python
# Function to handle the chatbot conversation
def chat_complete_messages(messages, temperature=0.7):
    completion = openai.ChatCompletion.create(
        model="gpt-4o",  # Or use "gpt-4o" if available for your use case
        messages=messages,
        temperature=0.7
    )
    return completion.choices[0].message['content']

# Function to query Pinecone for relevant info
def get_relevant_info(query, top_k=3):
    query_embedding = get_embeddings([query])[0]

    query_result = index.query(
        vector=query_embedding,
        top_k=top_k,  # Number of top results to return
        include_metadata=True
    )

    return query_result

# Main chatbot loop
while True:
    user_input = input("User Input: ")
    if user_input.lower() in ["exit", "bye"]:
        print("ChatBot: Goodbye! Have a great day!")
        break

    chatContext.append({'role': 'user', 'content': user_input})

    # Check if the user is asking about something stored in Pinecone
    if any(keyword in user_input.lower() for keyword in json_data.keys()):
        result = get_relevant_info(user_input)
        if result['matches']:
            response_message_content = "I found some relevant information:\n"
            for match in result['matches']:
                response_message_content += f"- {match['metadata']['text']} (from {match['metadata']['type']})\n"
        else:
            response_message_content = "I'm sorry, I couldn't find any relevant information."
    else:
        response_message_content = chat_complete_messages(chatContext, 0.7)

    print("ChatBot: ", response_message_content)

    chatContext.append({'role': 'assistant', 'content': response_message_content})
```
